```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = cv2.imread('eagle.jpg')

print(f"This image is type {type(image)} image dimensions {image.shape}")
```

```python
#save image to not affect the old version
image_copy = np.copy(image)

#change from color from BGR to RGB
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

#display the image
plt.imshow(image_copy)
```

### Define color threshold

```python
#define our color selection boundaries in rgb values
lower_blue = np.array([0, 0, 230])
higher_blue = np.array([50, 50, 250])
```

### Create Mask

```python
#Define the mask area
mask = cv2.inRange(image_copy, lower_blue, higher_blue)

#visualize the masked area
plt.imshow(mask, cmap='gray')
```

```python
#mask the image to let the bird show through
masked_image = np.copy(image_copy)
masked_image[mask != 0] = [0,0,0]
plt.imshow(masked_image)
```
