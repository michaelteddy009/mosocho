```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#make the y and x axis for the plot, the dev list are for different developers
ages_x = [x for x in range(25, 36)]
dev_y = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748,73752]
py_dev_y = [45372,48876,53850,57287,63016,65998,70003,70000,71496,75370,83640]
```

```python
#select a style to use for the pictorial use xkcd and see the change that happens
plt.style.use('dark_background')
#plt.xkcd()

#make the plot and use the show function to show the plot
plt.plot(ages_x, dev_y, linestyle='--', marker='.', linewidth=3, label='All_Developers')

plt.plot(ages_x, py_dev_y, color='red', marker='o', linewidth=3, label='Python')
#format_string_order = '[marker][linestyle][color]'

plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.tight_layout()  #to make padding for easier visualisation
plt.grid(True)  #to make a grid and False defaults to no grid

overalMediaSalary = 42856

plt.fill_between(ages_x, py_dev_y, dev_y, where=(py_dev_y > dev_y), interpolate=True, alpha=0.25, label='Above Avg')

plt.legend(['All_Developers', 'Python', 'Above Avg'], loc='upper left') #this adds labels inorder of how the lines were plotted but we still need to call legend once we have given the plots a label


plt.savefig('plots/plt.png') #to save the plt to a png file that we can use in our websites

plt.show()  #Displays the plotting
```
