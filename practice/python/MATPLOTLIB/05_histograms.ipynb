```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('dark_background')

ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
bins = [10, 20, 30, 40, 50, 60, ]

plt.hist(ages, bins=bins, edgecolor='black', color='yellow') #log means logrithmic scale to view the less data

plt.title('Ages of Respondant')
plt.xlabel('Ages')
plt.ylabel('Total Respondant')
plt.grid(False)

meanAge = sum(ages)/len(ages)
plt.axvline(meanAge, color='red', label='Mean Age', linewidth=3)

plt.tight_layout()
plt.legend()

plt.savefig('plots\histogram.png')

plt.show()
```
